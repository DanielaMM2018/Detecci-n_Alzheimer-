```python
Dataframe = Excel_Data.dropna( axis = 0)
print("Datos sin null = ", Dataframe.shape)
print("Datos con null = ", Excel_Data.shape)
```

```python
# Adecuar las etiquetas a binario 
#0 --> Paciente sano (CN), 1 --> pacientes con deterioro cognitivo leve (MCI) y pacientes con Alzheimer (EA)
l = Dataframe["Group"]
l = list(l)
grupo = []
for i in range(len(l)):
  if l[i] == 'CN':
    grupo.append(0)
  else:
    grupo.append(1)
Dataframe['Grupo'] = grupo
```

```python
Dataframe.head(67)
```

```python
#X = Excel_Data[,2:1602] #Características #[[:,0:1601]]
X = Dataframe.iloc[:,2:1602]
y = Dataframe['Grupo'].values #Etiqueta
```

```python
#Partición
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
#Normalización
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
pca = decomposition.PCA()
pca.fit(X_train)
X_trainPCA = pca.transform(X_train) 
l_1 = pca.explained_variance_ratio_
#l_1.append(pca.explained_variance_ratio_) 
print("Pesos de PCA:", l_1)
```

```python
suma = 0
cont = 0
for i in l_1:
  cont += 1
  suma += i
  if suma >= 0.98:
    print(cont)
    break
print(suma)
```

```python
pca = decomposition.PCA(n_components=127)
pca.fit(X_train)

PCA_X_train = pca.transform(X_train)
PCA_X_test = pca.transform(X_test)
PCA_X = pca.transform(X)
```

```python
def build_clf(unit):
# creating the layers of the NN
  ann = tf.keras.models.Sequential()
  ann.add(tf.keras.layers.Dense(units=unit, activation='relu'))
  ann.add(tf.keras.layers.Dense(units=unit, activation='relu'))
  ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
  ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
  return ann

model=KerasClassifier(build_fn=build_clf)

params={'batch_size':[100, 20, 50, 25, 32],
		'nb_epoch':[200, 100, 300, 400],
		'unit':[200, 300, 400, 500, 700, 1000],
		}
gs=GridSearchCV(estimator=model, param_grid=params, cv=10)
# now fit the dataset to the GridSearchCV object.
gs = gs.fit(PCA_X_train, y_train)


```

```python
#Accuracy
print("Clasificador: " + str(gs.best_estimator_) + " usando: " + str(gs.best_params_) + " - Accuracy: " + str(gs.best_score_))
```
